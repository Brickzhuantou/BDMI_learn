```python
import math
def solve_equ(b,c):
    val1 = b/2.0
    val2 = val1**2
    val3 = val2 + c
    val4 = math.sqrt(val3)
    root1 = val4 - val1
    root2 = -val4 - val1
    return root1, root2

print(solve_equ(10,39))
```

```python
# 归并排序的实现代码
B = [1,100,7,11,13,15,17,19,20,22]

def merge(A,B):
    i = 0
    j = 0
    C = []
    for t in range(len(A)+len(B)):
        if i < len(A) and j < len(B):
            if A[i] < B[j]:
                C.append(A[i])
                i = i+1
            else:
                C.append(B[j])
                j = j+1
        else:
            if i<len(A):
                C.append(A[i])
                i += 1
            if j<len(B):
                C.append(B[j])
                j += 1
    return C

# def merge(left, right):
#     '''合并操作，将两个有序数组left[]和right[]合并成一个大的有序数组'''
#     #left与right的下标指针
#     l, r = 0, 0
#     result = []
#     while l<len(left) and r<len(right):
#         if left[l] < right[r]:
#             result.append(left[l])
#             l += 1
#         else:
#             result.append(right[r])
#             r += 1
#     result += left[l:]
#     result += right[r:]
#     return result

def mergesort(A):
    n = len(A)
    
    if n <= 1:
        return A
    L = mergesort(A[:n//2])
    R = mergesort(A[n//2:])
    
    return merge(L, R)

            
print(mergesort(B))
```

```python
# 插入排序算法实现
def insert_sort(A):
    "插入排序"
    for i in range(1, len(A)):
        current = A[i]
        j = i-1
        while j >=0 and A[j] > current:
            A[j+1] = A[j]
            j -= 1
        A[j+1] = current

A = [2,3,1,0,0,1,9,1]
insert_sort(A)
print(A)
```
